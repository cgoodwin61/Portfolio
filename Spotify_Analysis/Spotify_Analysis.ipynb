If we use a cutoff of +/- .2, we see that there are a number of variables that have a pronounced impact on popularity: acousticness, energy, instrumentalness, loudness, speechiness, and year. Let's take a look at some scatter plots of these variables.

```python
cols = ['acousticness', 'energy', 'instrumentalness', 'loudness', 'speechiness', 'year']
```

Not surprisingly, these plots don't tell us much. But they are a good start to our modeling.

The first model I am going to look at is a RandomForestRegressor. Since popularity is a continuous variable, we cannot use a classifier method. I am going to first only use the columns we determined to be of greater importance, and then move on to using all available features. 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df[cols]
y = df['popularity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
train_scaled = scaler.fit_transform(X_train)
test_scaled = scaler.transform(X_test)
```

```python
rf = RandomForestRegressor()
rf.fit(train_scaled, y_train)
```

```python
preds = rf.predict(test_scaled)

accuracy = rf.score(test_scaled, y_test)
print("Test Accuracy: {:.4f}".format(accuracy*100))

average_error = (abs(y_test - preds)).mean()
print("{:.4f} average error".format(average_error))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
rf_mse = mean_squared_error(y_train, rf.predict(train_scaled))
rf_mae = mean_absolute_error(y_train, rf.predict(train_scaled))
```

```python
from math import sqrt
print("Random Forest training mse = ",rf_mse," & mae = ",rf_mae," & rmse = ", sqrt(rf_mse))
```

```python
rf_test_mse = mean_squared_error(y_test, rf.predict(test_scaled))
rf_test_mae = mean_absolute_error(y_test, rf.predict(test_scaled))
```

```python
print("Random Forest test mse = ",rf_test_mse," & mae = ",rf_test_mae," & rmse = ", sqrt(rf_test_mse))
```

Only a 62% accuracy on our test data isn't the best, but the average error is more indicitive of how our model is working. On the training data, we had a mean asbolute error of 3.5, meaning our predictions were only off by 3.5 on average. While the error increased to almost 9 on the test data, that still is not horrible given that popularity is on a scale from 1 to 100.
